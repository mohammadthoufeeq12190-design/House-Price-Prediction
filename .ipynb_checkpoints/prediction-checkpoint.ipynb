```python
import pandas as pd
import numpy as np
import pickle

# ============================
# Load Saved Model
# ============================
with open('rf_regression_model.sav', 'rb') as f:
    loaded_model = pickle.load(f)

print("Model Loaded Successfully!")
```

```python
# ============================
# User Inputs
# ============================
sqft = float(input("Enter Square Footage: "))
bedrooms = int(input("Enter Number of Bedrooms: "))
bathrooms = float(input("Enter Number of Bathrooms: "))
year_built = int(input("Enter Year Built: "))
lot_size = float(input("Enter Lot Size: "))
garage = int(input("Enter Garage Size (0/1/2...): "))
neigh_quality = int(input("Enter Neighborhood Quality (1-10): "))

# ============================
# Create Input DataFrame
# ============================
input_df = pd.DataFrame({
    "Square_Footage": [sqft],
    "Num_Bedrooms": [bedrooms],
    "Num_Bathrooms": [bathrooms],
    "Year_Built": [year_built],
    "Lot_Size": [lot_size],
    "Garage_Size": [garage],
    "Neighborhood_Quality": [neigh_quality]
})

# ============================
# Show Expected Features
# ============================
print("\nModel expects these features:")
print(loaded_model.feature_names_in_)

# ============================
# Predict
# ============================
predicted_price = loaded_model.predict(input_df)

predicted_value = round(float(predicted_price[0]), 2)

# ============================
# Output
# ============================
result_df = pd.DataFrame({
    "Square_Footage": [sqft],
    "Num_Bedrooms": [bedrooms],
    "Num_Bathrooms": [bathrooms],
    "Year_Built": [year_built],
    "Lot_Size": [lot_size],
    "Garage_Size": [garage],
    "Neighborhood_Quality": [neigh_quality],
    "Predicted_House_Price": [predicted_value]
})

print("\n=== House Price Prediction ===")
print(result_df)
```
